```python
data_copy = data_raw.copy(deep=True)
data_copy.isnull().sum()
```

```python
# print(list(data_copy['age'].apply(lambda x: np.nan if x=='?' else x)))
data_copy['age'] = data_copy['age'].apply(lambda x: np.nan if x=='?' else x)
data_copy['embarked'].fillna(data_copy['embarked'].mode()[0], inplace=True)
```

```python
data_copy.describe(include='all')
```

```python
data_copy.drop(columns=['cabin','boat','body','home.dest'], axis=1, inplace=True)
```

```python
data_copy.describe(include='all')
```

```python
(data_copy == '?').sum()
```
